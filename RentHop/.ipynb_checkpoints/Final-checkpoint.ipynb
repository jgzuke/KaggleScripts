```python
X_train = pd.read_json('data/train_leaky.json')
X_test = pd.read_json('data/test_leaky.json')
y_train = pd.read_json('data/train_interest.json')
scaler = StandardScaler()
train_X_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
test_X_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
#PCA
n_components = 3
pca_columns = ['layer1_pca_{}'.format(i) for i in range(n_components)]
pca = PCA(n_components=n_components)
pca_representation = pca.fit_transform(X_train)
pca_representation = pca.transform(X_test)

ridge = Ridge()
ridge.fit(train_X_scaled, y_train)
ridge_predictions = ridge.predict(test_X_scaled)

for col in pca_columns:
    X_test[col] = 0
X_test.loc[:, pca_columns] = pca_representation
X_test['layer1_ridge_predictions'] = ridge_predictions
```

```python
a = pd.read_csv('data/listing_image_time.csv')
b = pd.read_json('data/test_ids.json')
X_test['listing_id'] = b
a.columns = ['listing_id', 'time_stamp']
length = len(X_test)
X_test = X_test.merge(a, on='listing_id', how='outer')[:length]
```

```python
scaler = StandardScaler()

xtr = X_train
xtr = pd.DataFrame(scaler.fit_transform(xtr), columns=xtr.columns)
ytr = y_train
ytr_one_hot = np.eye(3)[ytr]
xte = X_test
xte = pd.DataFrame(scaler.fit_transform(xte), columns=xte.columns) # should be transform
```

```python
NFOLDS = 5
y_map = {'low': 2, 'medium': 1, 'high': 0}

params = {
    'eta':.01,
    'colsample_bytree':.6,
    'subsample':.6,
    'nthread':16,
    'objective':'multi:softprob',
    'eval_metric':'mlogloss',
    'num_class':3,
    'silent':1
}
```

```python
cols = list(X_train.columns)
dtrain = xgb.DMatrix(data=xtr[cols], label=ytr)
dtest = xgb.DMatrix(data=xte[cols])

bst = xgb.cv(params, dtrain, 10000, NFOLDS, early_stopping_rounds=50, verbose_eval=200)
best_rounds = np.argmin(bst['test-mlogloss-mean'])
print (bst['test-mlogloss-mean'][best_rounds])

bst = xgb.train(params, dtrain, best_rounds)
xgb_predictions = bst.predict(dtest)
```

```python
xgb_predictions = bst.predict(dtest)
listing_id = pd.read_json('data/test_ids.json').values
preds = pd.DataFrame(xgb_predictions)
preds.columns = ['high', 'medium', 'low']
preds['listing_id'] = listing_id
preds.to_csv('data/my_preds.csv', index=None)
```
