# AmericanSurvey

```python
%matplotlib inline
import pandas as pd
import numpy as np
columns=['SCHL', 'PINCP', 'NATIVITY', 'SEX']
population = pd.read_csv("data/pums/ss13pusa.csv",usecols=(columns), nrows=1000000)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
g = sns.FacetGrid(population, col='NATIVITY', hue="SEX", size=5)
g = (g.map(plt.scatter, "SCHL", "PINCP", edgecolor="w").add_legend())
```

```python
averages = np.array([[0]*4]*96)

for nativity in range(1,3):
    for sex in range(1,3):
        for schl in range(1,25):
            pincp_all = population[(population.SCHL == schl) & (population.NATIVITY == nativity) & (population.SEX == sex)]
            pincp = pincp_all.PINCP.values
            pincp = np.average(pincp[~np.isnan(pincp)]) if len(pincp) > 0 else 0
            index = schl + ((sex-1)*24) + ((nativity-1)*48) - 1
            averages[index] = [schl, pincp, nativity, sex]

averaged = pd.DataFrame(averages, columns=columns)
```

```python
g = sns.FacetGrid(population, col='NATIVITY', hue="SEX", size=5)
g = (g.map(sns.pointplot, "SCHL", "PINCP", scale=.7).add_legend())
```

```python
g = sns.FacetGrid(averaged, col='NATIVITY', hue="SEX", size=5)
g = (g.map(sns.regplot, "SCHL", "PINCP", scatter_kws={"s": 80}, order=2, truncate=True).add_legend())
```
